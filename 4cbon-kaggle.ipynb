## The 16-Layer Pipeline

```
L0 -> P -> W -> LX -> LA -> LC -> L1 -> L2 -> L3 -> L4 -> LR -> L6 -> L7 -> L8 -> L9 -> L10
```

### Zone A: Ingestion
| Layer | Name | Job |
|-------|------|-----|
| L0 | Interpretation Engine | Reads input. Infers intent. Injects prior self-beliefs. |
| P | Parsing Layer | Breaks answer into logical units. Lists claims and gaps. |
| W | World Model Layer | Labels every claim HIGH / MEDIUM / UNKNOWN certainty. |

### Zone B: Reality Pressure (New)
| Layer | Name | Job |
|-------|------|-----|
| LX | Reality Adjudication | Tests every claim: FALSIFIABLE, UNFALSIFIABLE, or TESTABLE-IN-PRINCIPLE |
| LA | Adversarial Countermodel | Actively tries to destroy the answer's core claims |
| LC | Compression Integrity | Finds where metaphor replaced mechanism, elegance erased uncertainty |

### Zone C: Arbitration
| Layer | Name | Job |
|-------|------|-----|
| L1 | Hypothesis Engine | Generates exactly 3 improvement hypotheses including a failure mode |
| L2 | Evaluation Layer | Scores each hypothesis 1-10. Picks best path. |
| L3 | Rewrite Planner | Plans the rewrite: what stays, changes, gets added, gets removed |

### Zone D: Execution
| Layer | Name | Job |
|-------|------|-----|
| L4 | Finalization Engine | Executes the rewrite. Produces the final improved answer. |

### Zone E: Reflection
| Layer | Name | Job |
|-------|------|-----|
| LR | Regret Layer | What errors corrected? Hallucinations removed? What still needs work? |
| L6 | Trace Memory | Immutable execution log. Append-only. The fossil record of every run. |
| L7 | Curriculum Generator | Lessons learned, failure patterns, reusable heuristics |
| L8 | Identity Model | Writes one self-belief about system behavior. Saved to memory. |
| L9 | Socratic Integrity Engine | Generates 3 self-questions specific to this run |
| L10 | Synthesis/Audit | Final certification: CERTIFIED / CERTIFIED WITH CAUTION / REQUIRES REVIEW |

### The Gap

This produced a 62→25 degradation — not from reasoning failure but from an upstream hypothesis selection error that propagated unchecked to L3.

---

### The Epistemic Shadow Layer

The same session produced a precise architectural term for what LX currently is in the pipeline.

ChatGPT proposed: **Epistemic Shadow Layer**

Definition: LX produces structured truth-status annotations per claim. It is fully visible to downstream layers including L10. It does not directly modify L4 execution inputs. It is present, observable, logged, but not yet mechanically influencing downstream decisions.

The deeper insight: LX already influences cognition indirectly by existing in the same context window. L4 reads LX output. The influence is indirect but real. The architectural tension is not whether LX is causal — it is whether the system is honest about how indirectly it already influences generation.

Three types of causality in the pipeline:
- **Hard causal** — execution (L4 runs the rewrite)
- **Soft causal** — guidance signals (compressed LX injected into L3)
- **Non-causal observational** — shadow layers (LX as currently implemented)

---

### The Fix: LP — Policy Translation Layer

Designed but not yet built. LP sits between L2 and L3.

Its job: check whether the hypothesis selected by L2 is structurally coherent with the original answer before allowing L3 to plan. If the hypothesis inverts the original's foundational claim, LP flags it and sends L1 back to regenerate.

This closes the L2-L3 gap.
